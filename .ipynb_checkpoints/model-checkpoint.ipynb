# Feature selection :- 

Feature selection is a crucial step in machine learning that involves selecting a subset of relevant features from a larger set of available features. The goal is to improve model performance by reducing dimensionality, eliminating irrelevant or redundant features, and enhancing interpretability.

### Spliting our dataset into fetature matrix  X and y

```python
X = df.drop('Default_payment',axis=1)
y = df['Default_payment']
```

```python
# univariate method.
from sklearn.feature_selection import SelectKBest,f_regression
selector = SelectKBest(score_func=f_regression, k=10)
X_selected = selector.fit_transform(X, y)
selected_indices = selector.get_support(indices=True)
# Get the selected feature names
selected_features = X.columns[selected_indices]
print("Selected Features:")
for feature in selected_features:
    print(feature)
```

## Outlier  Observation & treatment in above selected feature.

```python
mean = df['LIMIT_BAL'].mean()
std = df['LIMIT_BAL'].std()

        
lower_bound = mean - 3*std
upper_bound = mean + 3*std

df['LIMIT_BAL'] = np.where((df['LIMIT_BAL']<lower_bound) | (df['LIMIT_BAL']> upper_bound),df['LIMIT_BAL'].mean(),df['LIMIT_BAL'])
```

```python
mean = df['PAY_AMT1'].mean()
std = df['PAY_AMT1'].std()

        
lower_bound = mean - 3*std
upper_bound = mean + 3*std

df['PAY_AMT1'] = np.where((df['PAY_AMT1']<lower_bound) | (df['PAY_AMT1']> upper_bound),df['PAY_AMT1'].median(),df['PAY_AMT1'])
```

```python
mean = df['PAY_AMT2'].mean()
std = df['PAY_AMT2'].std()

        
lower_bound = mean - 3*std
upper_bound = mean + 3*std


df['PAY_AMT2'] = np.where((df['PAY_AMT2']<lower_bound) | (df['PAY_AMT2']> upper_bound),df['PAY_AMT2'].median(),df['PAY_AMT2'])
```

```python
mean = df['PAY_AMT4'].mean()
std = df['PAY_AMT4'].std()

        
lower_bound = mean - 3*std
upper_bound = mean + 3*std

df['PAY_AMT4'] = np.where((df['PAY_AMT4']<lower_bound) | (df['PAY_AMT4']> upper_bound),df['PAY_AMT4'].median(),df['PAY_AMT4'])
```

```python
X = df[['LIMIT_BAL','PAY_1','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6','PAY_AMT1','PAY_AMT2','PAY_AMT4']]
```

```python
X.head()
```

### Splitting our feature matrix  into train and test set.

```python
from sklearn.model_selection import  train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=42)
```

```python
df['Default_payment'].value_counts()
```

Observation:- Our Target feature( y= Default) is highly imbalanced.

## Imbalance class treatment

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
X_train_ros,y_train_ros = ros.fit_resample(X_train,y_train)
print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes after fit {}".format(Counter(y_train_ros)))
```

**RandomOverSampler**:-

Over-sample the minority class(es) by picking samples at random with replacement.

## Model Building

```python
# Import the required libraries
from sklearn.ensemble import VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier

# Create individual classification models
model1 = DecisionTreeClassifier()
model2 = RandomForestClassifier()
model3 = AdaBoostClassifier()
model4 = SVC()
model5 = LogisticRegression()
model6 = RidgeClassifier()

# Create the ensemble model using VotingClassifier
ensemble_model = VotingClassifier(estimators=[('dt', model1), ('rf', model2),('ada',model3), ('svc', model4),('lr',model5),('lidge',model6)], voting='hard')

# Fit the ensemble model on the training data
ensemble_model.fit(X_train_ros, y_train_ros)

# Predict the target variable using the ensemble model
y_pred = ensemble_model.predict(X_test)
```

**Voting classifier**:- 

A voting classifier is a machine learning algorithm that combines the predictions of multiple individual classifiers to make a final prediction.In a voting classifier, each individual classifier is trained on the same dataset using different algorithms or configurations. These classifiers can be of different types, such as decision trees, support vector machines, or neural networks. Each classifier independently makes a prediction, and the final prediction is determined by a majority vote or by taking the average of the individual predictions.

```python
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
train_pred = ensemble_model.predict(X_train_ros)
test_pred = ensemble_model.predict(X_test)
train_accuracy = accuracy_score(y_train_ros, train_pred)
test_accuracy = accuracy_score(y_test, test_pred)
print(f"Model: {ensemble_model.__class__.__name__}")
print(f"Train Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")
print(classification_report(y_test, test_pred))
print(confusion_matrix(y_test, test_pred))
```

```python
sns.heatmap(confusion_matrix(y_test, test_pred),annot=True,xticklabels=[0,1],yticklabels=[0,1],cmap='viridis')
```
